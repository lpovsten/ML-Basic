```python
#### 5. Зависимость ошибки от ранга  - генерация матрицы и построение графика 
#import sys
#!{sys.executable} -m pip install numpy matplotlib

import numpy as np
import matplotlib.pyplot as plt

n = 100
A = np.random.randn(n, n)

U, S, Vt = np.linalg.svd(A, full_matrices=False)
S_diag = np.diag(S)

errors = []
for r in range(2, n+1):
    # Аппроксимация
    A_r = U[:, :r] @ S_diag[:r, :r] @ Vt[:r, :]
    # Ошибка
    error = np.linalg.norm(A - A_r, 'fro')
    errors.append(error)

plt.plot(range(2, n+1), errors)
plt.xlabel('Ранг r')
plt.ylabel('Ошибка E(r)')
plt.title('Зависимость ошибки от ранга')
plt.grid(True)
plt.show()
```

# Задача 1: Градиент функции
## Посчитать градиент функции:
$f(x) = x_1^3 - 2x_1x_2 + x_2^2 - 3x_1 - 2x_2$

---
## Решение

## 1. Вычисляем градиент

Градиент функции — вектор из частных производных:

$\nabla f(x_1, x_2) = \begin{pmatrix} \frac{\partial f}{\partial x_1} \\ \frac{\partial f}{\partial x_2} \end{pmatrix}$

### 1.1 Производная по $x_1$:

$
\begin{aligned}
\frac{\partial f}{\partial x_1} &= \frac{\partial}{\partial x_1}(x_1^3) + \frac{\partial}{\partial x_1}(-2x_1x_2) + \frac{\partial}{\partial x_1}(x_2^2) + \frac{\partial}{\partial x_1}(-3x_1) + \frac{\partial}{\partial x_1}(-2x_2) \\
&= 3x_1^2 - 2x_2 - 3
\end{aligned}
$

### 1.2 Производная по $x_2$:

$
\begin{aligned}
\frac{\partial f}{\partial x_2} &= \frac{\partial}{\partial x_2}(x_1^3) + \frac{\partial}{\partial x_2}(-2x_1x_2) + \frac{\partial}{\partial x_2}(x_2^2) + \frac{\partial}{\partial x_2}(-3x_1) + \frac{\partial}{\partial x_2}(-2x_2) \\
&= -2x_1 + 2x_2 - 2
\end{aligned}
$

### 1.3 Полный градиент:

$\boxed{\nabla f(x_1, x_2) = \begin{pmatrix} 3x_1^2 - 2x_2 - 3 \\ -2x_1 + 2x_2 - 2 \end{pmatrix}}$

## 2. Находим критические точки
 
Критические точки — решения системы $\nabla f(x_c) = 0$:
 
 $
 \begin{cases}
 3x_1^2 - 2x_2 - 3 = 0 \quad \text{(1)} \\
 -2x_1 + 2x_2 - 2 = 0 \quad \text{(2)}
 \end{cases}
 $
 
 ### 2.1 Из уравнения (2):
 
 $
 -2x_1 + 2x_2 - 2 = 0 \implies 2x_2 = 2x_1 + 2 \implies \boxed{x_2 = x_1 + 1}
 $
 
 ### 2.2 Подставляем в уравнение (1):
 
 $
 3x_1^2 - 2(x_1 + 1) - 3 = 0
 $
 
 $
 3x_1^2 - 2x_1 - 2 - 3 = 0
 $
 
 $
 3x_1^2 - 2x_1 - 5 = 0
 $
 
 ### 2.3 Решаем квадратное уравнение:
 
 Дискриминант:
 $D = (-2)^2 - 4 \cdot 3 \cdot (-5) = 4 + 60 = 64$$
 
 Корни:
 $x_1 = \frac{2 \pm \sqrt{64}}{2 \cdot 3} = \frac{2 \pm 8}{6}$
 
 Первое решение:
 $x_1^{(1)} = \frac{2 + 8}{6} = \frac{10}{6} = \frac{5}{3}$
 $x_2^{(1)} = x_1^{(1)} + 1 = \frac{5}{3} + 1 = \frac{5}{3} + \frac{3}{3} = \frac{8}{3}$
 
 Второе решение:
 $x_1^{(2)} = \frac{2 - 8}{6} = \frac{-6}{6} = -1$
 $x_2^{(2)} = x_1^{(2)} + 1 = -1 + 1 = 0$

## Ответ
 
 ### Градиент функции:
 
 $\boxed{\nabla f(x_1, x_2) = \begin{pmatrix} 3x_1^2 - 2x_2 - 3 \\ -2x_1 + 2x_2 - 2 \end{pmatrix}}$
 
 ### Критические точки:
 
 $\boxed{x_c^{(1)} = \left( \frac{5}{3}, \frac{8}{3} \right)}$
 
 $\boxed{x_c^{(2)} = (-1, 0)}$

# Задача 2: Проверить тождество


Функция  
$f = \ln(\sqrt{x_1} + \sqrt{x_2})$

Проверить: 
$x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{1}{2}$

---

## Решение

### 1. Найдём частные производные
Обозначим внутреннюю функцию 
$u = \sqrt{x_1} + \sqrt{x_2}$

тогда 
$f = \ln(u)$

$\frac{d}{du}[\ln(u)] = \frac{1}{u}$

$\frac{\partial f}{\partial u} = \frac{1}{u} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}}$


#### Производная по $x_1$:

$\frac{\partial f}{\partial x_1} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{\partial}{\partial x_1}(\sqrt{x_1} + \sqrt{x_2})$

$\frac{\partial}{\partial x_1}(\sqrt{x_1}) = \frac{1}{2\sqrt{x_1}}$  
$\frac{\partial}{\partial x_1}(\sqrt{x_2}) = 0$

Итого:
$\frac{\partial f}{\partial x_1} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_1}}$



#### Производная по $x_2$:

$\frac{\partial f}{\partial x_2} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{\partial}{\partial x_2}(\sqrt{x_1} + \sqrt{x_2})$

$\frac{\partial}{\partial x_2}(\sqrt{x_1}) = 0$  
$\frac{\partial}{\partial x_2}(\sqrt{x_2}) = \frac{1}{2\sqrt{x_2}}$

Итого:
$\frac{\partial f}{\partial x_2} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_2}}$



### 2. Подставим в выражение

$x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = x_1 \cdot \frac{1}{2\sqrt{x_1}(\sqrt{x_1} + \sqrt{x_2})} + x_2 \cdot \frac{1}{2\sqrt{x_2}(\sqrt{x_1} + \sqrt{x_2})}$



### 3. Упростим

Первое слагаемое:
$x_1 \cdot \frac{1}{2\sqrt{x_1}(\sqrt{x_1} + \sqrt{x_2})} = \frac{\sqrt{x_1}}{2(\sqrt{x_1} + \sqrt{x_2})}$

(т.к. $\frac{x_1}{\sqrt{x_1}} = \sqrt{x_1}$)

Второе слагаемое:  
$x_2 \cdot \frac{1}{2\sqrt{x_2}(\sqrt{x_1} + \sqrt{x_2})} = \frac{\sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})}$



### 4. Сложим

$x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{\sqrt{x_1}}{2(\sqrt{x_1} + \sqrt{x_2})} + \frac{\sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})}$

$= \frac{\sqrt{x_1} + \sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})}$



### 5. Получаем

$x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{1}{2}$



## Ответ

### Утверждение доказано:

$\boxed{x_1 \frac{\partial f}{\partial x_1} + x_2 \frac{\partial f}{\partial x_2} = \frac{1}{2}}$


# Задача 3: Матрица Якоби

Функция $f: \mathbb{R}^3 \to \mathbb{R}^2$:
$
f(x, y, z) = 
\begin{pmatrix}
x + y + z \\
xyz
\end{pmatrix}
$

Найти:
1. Матрицу Якоби $J_f$ функции $f$
2. Численное значение в точке $v = (1, 2, 3)^T$

---

## Решение

### 1. Определение матрицы Якоби

Для функции $f: \mathbb{R}^n \to \mathbb{R}^m$ матрица Якоби имеет размер $m \times n$:

$
J_f = 
\begin{bmatrix}
\frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} & \frac{\partial f_1}{\partial z} \\
\frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y} & \frac{\partial f_2}{\partial z}
\end{bmatrix}
$

Где:
- $f_1(x, y, z) = x + y + z$
- $f_2(x, y, z) = xyz$



### 2. Вычисляем частные производные

#### Для $f_1$:
$
\frac{\partial f_1}{\partial x} = 1, \quad
\frac{\partial f_1}{\partial y} = 1, \quad
\frac{\partial f_1}{\partial z} = 1
$

#### Для $f_2$:
$
\frac{\partial f_2}{\partial x} = yz, \quad
\frac{\partial f_2}{\partial y} = xz, \quad
\frac{\partial f_2}{\partial z} = xy
$



### 3. Матрица Якоби в общем виде
$
J_f(x, y, z) = 
\begin{bmatrix}
1 & 1 & 1 \\
yz & xz & xy
\end{bmatrix}
$



### 4. Численное значение в точке $v = (1, 2, 3)^T$
#### Подставляем $x = 1$, $y = 2$, $z = 3$:
$$ J_f(1, 2, 3) = 
\begin{bmatrix}
1 & 1 & 1 \\
2 \cdot 3 & 1 \cdot 3 & 1 \cdot 2
\end{bmatrix}
= 
\begin{bmatrix}
1 & 1 & 1 \\
6 & 3 & 2
\end{bmatrix}
$$



## Ответ
1. Матрица Якоби:
$
\boxed{J_f(x, y, z) = 
\begin{bmatrix}
1 & 1 & 1 \\
yz & xz & xy
\end{bmatrix}}
$

2. Значение в точке $v = (1, 2, 3)^T$:
$
\boxed{J_f(1, 2, 3) = 
\begin{bmatrix}
1 & 1 & 1 \\
6 & 3 & 2
\end{bmatrix}}
$

# Задача 4: Куб евклидовой нормы

Функция

$
f(x) = \frac{1}{3} \|x\|_2^3, 
$
где $x \in \mathbb{R}^n$ \ {0}

Найти:
1. Первый дифференциал $df(x)$
2. Градиент $\nabla f(x)$

---

## Решение

### 1. Обозначим евклидову норму

Пусть $r = \|x\|_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}$.

Тогда:
$
f(x) = \frac{1}{3} r^3
$

$r = (x_1^2 + x_2^2 + \dots + x_n^2)^{1/2}$

### 2. Найдём производную нормы

$\frac{\partial r}{\partial x_i} = \frac{1}{2}(x_1^2 + \dots + x_n^2)^{-1/2} \cdot 2x_i$

$\frac{\partial r}{\partial x_i} = \frac{x_i}{\sqrt{x_1^2 + \dots + x_n^2}} = \frac{x_i}{r}$

В векторной форме:
$
\nabla r = \frac{x}{r}
$

### 3. Градиент функции $f$

Используем цепное правило:
$
\nabla f(x) = \frac{d}{dr}\left(\frac{1}{3}r^3\right) \cdot \nabla r
$

$
\frac{d}{dr}\left(\frac{1}{3}r^3\right) = \frac{1}{3} \cdot \frac{d}{dr}(r^3) = r^2
$

Следовательно:
$
\nabla f(x) = r^2 \cdot \frac{x}{r} = r \cdot x
$

Подставляем $r = \|x\|_2$:
$
\boxed{\nabla f(x) = \|x\|_2 \cdot x}
$


### 4. Первый дифференциал

Первый дифференциал связан с градиентом:
$
df(x) = \nabla f(x)^T dx
$

Таким образом:
$
\boxed{df(x) = \|x\|_2 \cdot x^T dx}
$



## Ответ

1. Градиент:
$
\boxed{\nabla f(x) = \|x\|_2 \cdot x}
$

2. Первый дифференциал:
$
\boxed{df(x) = \|x\|_2 \cdot x^T dx}
$

# Задача 5: Евклидова норма

Функция $
f(x) = \|x\|_2, $
где $x \in \mathbb{R}^n$ \ {0}


Найти:
1. Первый дифференциал $df(x)$
2. Градиент $\nabla f(x)$

---

## Решение

### 1. Обозначим норму
Пусть $r = \|x\|_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}$

Тогда:
$
f(x) = r
$

### 2. Градиент нормы (из задачи 4)
$
\frac{\partial r}{\partial x_i} = \frac{x_i}{r}, \quad i = 1, \dots, n
$

В векторной форме:
$
\nabla r = \frac{x}{r}
$

Поскольку $f(x) = r$, то:
$
\boxed{\nabla f(x) = \frac{x}{\|x\|_2}}
$

### 3. Первый дифференциал
Первый дифференциал связан с градиентом:
$
df(x) = \nabla f(x)^T dx
$

Где $dx = (dx_1, dx_2, \dots, dx_n)^T$ — вектор дифференциалов координат.

Подставляем:
$
df(x) = \frac{x^T}{\|x\|_2} dx
$


## Ответ

1. Градиент:
$
\boxed{\nabla f(x) = \frac{x}{\|x\|_2}}
$

2. Первый дифференциал:
$
\boxed{df(x) = \frac{x^T}{\|x\|_2} dx}
$
